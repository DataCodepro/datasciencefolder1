```python
import numpy as np
import pandas as pd
import requests
```

```python
url_link = 'https://finance.yahoo.com/quote/DOT-USD/history?p=DOT-USD'
r = requests.get(url_link,headers ={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'})
read_html_pandas_data = pd.read_html(r.text)[0]
df= pd.DataFrame(read_html_pandas_data)
df.head()
```

```python
df.tail()
```

```python
df.drop(100,axis=0,inplace =True)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.to_csv('dotcsv.csv',index =False)
```

```python
df.to_excel('dotexcel.xlsx',index =False)
```

```python
with pd.ExcelWriter('dottranc.xlsx',engine= 'xlsxwriter') as writer:
    df[(df['Date']>=pd.Timestamp(2022,6,1))&(df['Date']<=pd.Timestamp(2022,6,30))].to_excel(writer,sheet_name='JUNEREC',index=False,engine = 'xlsxwriter')
    df[(df['Date']>=pd.Timestamp(2022,7,1))&(df['Date']<=pd.Timestamp(2022,7,31))].to_excel(writer,sheet_name='JULYREC',index=False,engine = 'xlsxwriter')
    df[(df['Date']>=pd.Timestamp(2022,8,1))&(df['Date']<=pd.Timestamp(2022,8,31))].to_excel(writer,sheet_name='AUGREC',index=False,engine = 'xlsxwriter')
```
