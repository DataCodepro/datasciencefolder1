```python
#rename will change row and columns name where axis set 1 is column and axis set 0 is row
df.rename({'Date':'DATE', 'Region':'REGION', 'Type':'TYPE', 'Units':'UNITS', 'Sales':'SALES'},axis =1,inplace =True)
df.head()
```

```python
#groupy is used when an agg fun is to be applied on a column(categorcal)
df.groupby('REGION').apply('sum')
```

```python
pv = pd.pivot_table(df,index='REGION',columns='TYPE',values=['UNITS','SALES'],aggfunc=max,margins=True)
pv 
```

```python
pv.to_csv('pivot.csv',index=True)
```

```python
#CROSS TABULATION CHECKING THE CORRELATION BETWEEN TWO COLUMNS
pd.crosstab(df['REGION'],df['TYPE'],margins=True)
```

```python
#CONDITIONAL SELECTION OF VALUES
df[df['REGION']== 'East']
```

```python
df[(df['REGION']=='West') & (df['SALES'] >= 500) &(df['UNITS']<=20)]
```
