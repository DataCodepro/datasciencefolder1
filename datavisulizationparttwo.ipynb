```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_tip = sns.load_dataset('tips')
df_tip.head()
```

```python
sns.displot(x = df_tip.total_bill,kde =True)
```

```python
sns.regplot(x=df_tip.total_bill,y = df_tip.tip)
```

```python
sns.regplot(x=df_tip.total_bill,y = df_tip['size'])
```

```python
sns.jointplot(x=df_tip['total_bill'],y=df_tip['tip'])
```

```python
sns.pairplot(df_tip)
```

```python
sns.pairplot(df_tip,hue ='sex')
```

```python
sns.countplot(x= df_tip['time'])
```

```python
sns.barplot(x=df_tip['sex'],y=df_tip['total_bill'],estimator=max)
```

```python
sns.barplot(x=df_tip['sex'],y=df_tip['total_bill'],estimator=max,hue = df_tip['time'],palette='gist_earth')
```

```python
sns.barplot(x=df_tip['sex'],y=df_tip['total_bill'],estimator=sum,hue = df_tip['time'],palette='gist_earth')
```

```python
sns.violinplot(x=df_tip['sex'],y=df_tip['total_bill'],estimator=sum,hue = df_tip['time'],palette='gist_earth')
```

```python
sns.swarmplot(x=df_tip['sex'],y=df_tip['total_bill'],hue = df_tip['time'],palette='gist_earth')
```

```python
sns.stripplot(x=df_tip['sex'],y=df_tip['total_bill'],hue = df_tip['time'],palette='gist_earth')
```

```python
sns.boxplot(x= df_tip['day'],y=df_tip['total_bill'])
```

```python
df_flight = sns.load_dataset('flights')
df_flight
```

```python
df = pd.pivot_table(df_flight,values='passengers',index='year',columns='month')
df
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(df,annot =True,fmt='0.5g')
```

```python
plt.figure(figsize=(8,8))
sns.clustermap(df,annot =True,fmt='0.5g')
```
